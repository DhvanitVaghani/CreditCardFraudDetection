```python
# Import pandas for data manipulation
import pandas as pd

# Load the Dataset and convert it into Dataframe
credit_card_df = pd.read_csv('creditcard.csv')

# Display the first 10 records of the dataset
credit_card_df.head(10)
```

```python
# Import seaborn for data visulization
import seaborn as seab

# Import matplotlib for data visulization
import matplotlib.pyplot as plt

# Show histogram for label distribution
seab.histplot(credit_card_df['Class'])
```

```python
# Import train_test_split for splitting training and testing
from sklearn.model_selection import train_test_split

# Assign features and label into separate variables
features = credit_card_df.drop('Class', axis=1)
label = credit_card_df.loc[:,['Class']]

# Split data into training and testing
train_features, test_features, train_label, test_label = train_test_split(features, label, test_size=0.3)
```

```python
# Import StandardScaler for scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Scale train and test features
train_features = scaler.fit_transform(train_features)
test_features = scaler.transform(test_features)
```

```python
test_features
```

```python
# Import all metrics for classification evalution
from sklearn.metrics import roc_auc_score, accuracy_score, f1_score, recall_score, precision_score, confusion_matrix

# Import display module for showing output nicely
from IPython.display import display

# Show all the required metrics
def show_evalution_table(model_name,true_label,prediction):
    accuracy = accuracy_score(true_label,prediction)
    tn, fp, fn, tp = confusion_matrix(true_label,prediction).ravel()
    tpr = tp/(tp+fn)
    tnr = tn/(tn+fp)
    ppv = tp/(tp+fp)
    npv = tn/(tn+fn)
    f1 = f1_score(true_label,prediction)
    roc_auc = roc_auc_score(true_label,prediction)
    index = ['Accuracy','TPR (Sensitivity, Recall)','TNR (Specificity)','PPV (Precision)','NPV','F1-Measure','ROC/AUC Curves']
    evalution_df = pd.DataFrame({model_name: pd.Series([accuracy*100,tpr*100,tnr*100,ppv*100,npv*100,f1*100,roc_auc*100],index=index)})
    display(evalution_df)
```

```python
# Trainning and Testing using KNN

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)

# Fit features and label to the model
knn.fit(train_features, train_label.values.ravel())

# Make predictions
prediction = knn.predict(test_features)

# Show Evalution table
show_evalution_table('KNN',test_label,prediction)
```

```python
# Show Confusion Matrix
def show_confusion_matrix(true_label,prediction):
    matrix = confusion_matrix(true_label, prediction)

    plt.figure(figsize=(3, 2))
    seab.heatmap(matrix, annot=True, fmt='d', cmap='Reds')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()
```

```python
show_confusion_matrix(test_label,prediction)
```

```python
# Use k = 3 and train and test the model again
knn.set_params(n_neighbors=3)

# Fit features and label to the model
knn.fit(train_features, train_label.values.ravel())

# Make predictions
prediction = knn.predict(test_features)

# Show Evalution table
show_evalution_table('KNN',test_label,prediction)
```

```python
show_confusion_matrix(test_label,prediction)
```

```python
# Use SMOTE method for balancing data
from imblearn.over_sampling import SMOTE

sm = SMOTE()

smote_featurs, smote_label = sm.fit_resample(features,label)

featurs = pd.DataFrame(data=smote_featurs,columns=features.columns)
label = pd.DataFrame(data=smote_label,columns=label.columns)
```

```python
seab.histplot(label)
```

```python
train_features, test_features, train_label, test_label = train_test_split(featurs, label, test_size=0.3)
```

```python
train_features.values
```

```python
# Train and Test after applying SMOTE
knn.fit(train_features.values, train_label.values.ravel())

# Make predictions
prediction = knn.predict(test_features.values)

# Show Evalution table
show_evalution_table('KNN',test_label,prediction)

show_confusion_matrix(test_label,prediction)
```

```python
# Trainning and Testing using Decision Tree
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()

# Fit features and label to the model
dtc.fit(train_features.values, train_label.values.ravel())

# Make predictions
prediction = dtc.predict(test_features.values)

# Show Evalution table
show_evalution_table('Decision Tree',test_label,prediction)
```

```python
show_confusion_matrix(test_label,prediction)
```

```python
# Trainning and Testing using Logistic Regression
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

# Fit features and label to the model
lr.fit(train_features.values, train_label.values.ravel())

# Make predictions
prediction = lr.predict(test_features.values)

# Show Evaluations matrix
show_evalution_table('Logistic Regression',test_label,prediction)
show_confusion_matrix(test_label,prediction)
```

```python
# Trainning and Testing using Random Forest
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()

# Fit features and label to the model
rfc.fit(train_features.values, train_label.values.ravel())

# Show Evaluations matrix
prediction = rfc.predict(test_features.values)

# Make predictions
show_evalution_table('Random Forest',test_label,prediction)
show_confusion_matrix(test_label,prediction)
```
